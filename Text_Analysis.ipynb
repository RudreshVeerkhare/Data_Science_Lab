```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

class DataPoint:
    fulldata = []
    def __init__(self, sentence, spam):
        self.text = sentence
        self.spam = spam
        self.tokens = word_tokenize(sentence)
        self.wip_text = None
        self.wip_tokens = None
    

tokens = []
for mail, spam in zip(df["text"], df["spam"]):
    DataPoint.fulldata.append(DataPoint(mail, spam))
```

### Preparing the Corpus

#### Lower case all words

```python
for dp in DataPoint.fulldata:
    dp.wip_text = dp.text.lower()

DataPoint.fulldata[0].wip_text
```

#### Remove all punctuations

```python
from nltk.tokenize import RegexpTokenizer
import string

# tokenize wip_text first
tokenizer = RegexpTokenizer(r'[a-zA-Z0-9]+')
for dp in DataPoint.fulldata:
    dp.wip_tokens = tokenizer.tokenize(dp.wip_text)

DataPoint.fulldata[0].wip_tokens
```

#### remove all english stopwords

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stopwords = set(stopwords.words('english'))

for dp in DataPoint.fulldata:
    dp.wip_tokens = list(filter(lambda x: x not in stopwords, dp.wip_tokens))

DataPoint.fulldata[0].wip_tokens
```

#### Stem words in corpus

```python
from nltk.stem import SnowballStemmer
sno = SnowballStemmer("english")

for dp in DataPoint.fulldata:
    dp.wip_tokens = list(map(lambda x: sno.stem(x), dp.wip_tokens))

DataPoint.fulldata[0].wip_tokens
```

#### Build a document term matrix from the corpus, called dtm.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer 

def fn_tdm_df(docs, xColNames = None, **kwargs):
    ''' create a term document matrix as pandas DataFrame
    with **kwargs you can pass arguments of CountVectorizer
    if xColNames is given the dataframe gets columns Names'''

    #initialize the  vectorizer
    vectorizer = CountVectorizer(**kwargs)
    x1 = vectorizer.fit_transform(docs)
    #create dataFrame
    _df = pd.DataFrame(x1.toarray().transpose(), index = vectorizer.get_feature_names())
    if xColNames is not None:
        _df.columns = xColNames

    return _df

dtm = fn_tdm_df(
    map(lambda x: " ".join(x.wip_tokens), DataPoint.fulldata)
)
```

#### How many terms are in dtm?

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer 

def fn_tdm_df(docs, xColNames = None, **kwargs):
    ''' create a term document matrix as pandas DataFrame
    with **kwargs you can pass arguments of CountVectorizer
    if xColNames is given the dataframe gets columns Names'''

    #initialize the  vectorizer
    vectorizer = CountVectorizer(**kwargs)
    x1 = vectorizer.fit_transform(docs)
    #create dataFrame
    _df = pd.DataFrame(x1.toarray().transpose(), index = vectorizer.get_feature_names())
    if xColNames is not None:
        _df.columns = xColNames

    return _df

dtm = fn_tdm_df(
    map(lambda x: " ".join(x.wip_tokens), DataPoint.fulldata), min_df = 0.05
)
```

```python
dtm
```

### Building machine learning models

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dtm.T, list(map(lambda x: x.spam, DataPoint.fulldata)), test_size=0.30, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
y_pred_train = clf.predict(X_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
PAD = 25
print("=" * PAD , "Train" , "=" * PAD)
print(classification_report(y_train, y_pred_train, target_names=["Ham", "Spam"]), end = "\n\n\n")

print("=" * PAD , "Test" , "=" * PAD)
print(classification_report(y_test, y_pred, target_names=["Ham", "Spam"]))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
y_pred_train_rf = model.predict(X_train)
y_pred_rf = model.predict(X_test)
```

```python
PAD = 25
print("=" * PAD , "Train" , "=" * PAD)
print(classification_report(y_train, y_pred_train_rf, target_names=["Ham", "Spam"]), end = "\n\n\n")

print("=" * PAD , "Test" , "=" * PAD)
print(classification_report(y_test, y_pred_rf, target_names=["Ham", "Spam"]))
```
